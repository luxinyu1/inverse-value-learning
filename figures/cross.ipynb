```python
import matplotlib.pyplot as plt
import seaborn as sns

# 设置Seaborn样式
sns.set(style="whitegrid")

# 准备数据
values = [7.23, 3.85, 4.79, 5.16]
models = [
    'Llama3-8B\nFull Finetuning',
    'Llama3-8B-Base+\nTinyLlama 1.1B curriculum',
    'Llama3-8B-Base+\nLlama-3.2 1B',
    'Llama3-8B-Base+\nLlama-3.2 1B norm'
]

palette = sns.color_palette("Blues", len(values))

sorted_palette = []
palette = sns.color_palette("Blues", len(values))
sorted_palette.append(palette[0])
sorted_palette.append(palette[1])
sorted_palette.append(palette[2])
sorted_palette.append(palette[3])

fig, ax = plt.subplots(figsize=(6, 6))

sns.barplot(x=models, y=values, palette=sorted_palette, ax=ax, width=0.65)

plt.xticks(rotation=45, ha='right', fontsize=13)

ax.set_xlabel('Settings', fontsize=14)
ax.set_ylabel('MT-Bench Score', fontsize=14)

ax.tick_params(axis='x', colors='black')
ax.tick_params(axis='y', colors='black')

for i, value in enumerate(values):
    ax.text(i, value + 0.1, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(True, axis='y', linestyle='--', linewidth=1, alpha=0.65)

# 显示图表
plt.tight_layout()
plt.savefig('cross.pdf', format='pdf')
plt.show()
```
