```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
models = ['1.1B', '1.1B', '7B', '7B', '13B', '13B']
settings = ['Linear Probe', 'Full Finetuning', 'Linear Probe', 'Full Finetuning', 'Linear Probe', 'Full Finetuning']
values = [1.36, 3.84, 2.49, 5.93, 2.82, 6.23]

# Create a DataFrame for easy plotting
df = pd.DataFrame({
    'Model': models,
    'Setting': settings,
    'Score': values
})

print(df)

# Set up the color palette
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")

# Create the barplot
plt.figure(figsize=(6, 6))
ax = sns.barplot(x='Model', y='Score', hue='Setting', order=['1.1B', '7B', '13B'], hue_order=['Linear Probe', 'Full Finetuning'], data=df, palette=palette)

hatches = ['/', '/', '/' , '\\', '\\', '\\', '/', '\\']
bars = ax.patches

for i, bar in enumerate(bars):
    bar.set_hatch(hatches[i])  # Cycle through hatches

# Add titles and labels
plt.xlabel('# Parameters', fontsize=16)
plt.ylabel('MT-Bench Score', fontsize=16)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Increase legend font size and adjust position
plt.legend(fontsize=16, loc='upper left')

plt.tight_layout()

plt.savefig('preliminary.pdf', format='pdf')
plt.show()
plt.close()
```
