```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
models = ['1.1B+1.1B', '1.1B+1.1B', '1.1B+1.1B', '7B+1.1B', '7B+1.1B', '7B+1.1B']
settings = ['Cascade+', 'Cascade', 'Full Residual', 'Cascade+', 'Cascade', 'Full Residual']
values = [3.89, 2.75, 3.62, 4.45, 4.2, 4.62]

# Create a DataFrame for easy plotting
df = pd.DataFrame({
    'Model': models,
    'Setting': settings,
    'Score': values
})

# Set up the color palette
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")

print(len(palette))

# Create the barplot
plt.figure(figsize=(4, 6))
ax = sns.barplot(x='Model', y='Score', hue='Setting', data=df, palette=palette)

hatches = ['/', '/', '\\', '\\', 'x', 'x', '/', '\\', 'x']  # List of hatch patterns
bars = ax.patches

for i, bar in enumerate(bars):
    bar.set_hatch(hatches[i % len(hatches)])  # Cycle through hatches

# Add titles and labels
plt.xlabel('Settings', fontsize=16)
plt.ylabel('MT-Bench Score', fontsize=16)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

ax.yaxis.set_label_position("right")
ax.yaxis.tick_right()

# Increase legend font size and adjust position
ax.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",
          mode="expand", borderaxespad=0, ncol=2, fontsize=12)

plt.tight_layout()

plt.savefig('arch.pdf', format='pdf')
plt.show()
plt.close()
```
